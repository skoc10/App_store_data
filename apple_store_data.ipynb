# Exploring app genres

```python
num_genre=len(appdf['prime_genre'].unique())
#appdf['prime_genre'].value_counts().count()  
print('Number of prime genre: ', num_genre)
```

```python
num_app_gen= appdf['prime_genre'].value_counts().sort_values(ascending=False)
num_app_gen.plot(kind='bar')
plt.figure(figsize=(10,10))
plt.show()
```

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go

num_app_gen= appdf['prime_genre'].value_counts().sort_values(ascending=False)
data = [go.Bar(
        x = num_app_gen.index, # index = category name
        y = num_app_gen.values, # value = count
)]

plotly.offline.iplot(data)
```

```python
app_free_num= appdf[ appdf['price']==0]
nag=app_free_num.prime_genre.value_counts()
data = [go.Bar(
        x = nag.index,
        y = num_app_gen.values
)]

plotly.offline.iplot(data)
```

```python
app_free_num= appdf[ appdf['price']>10]
nag=app_free_num.prime_genre.value_counts()
data = [go.Bar(
        x = nag.index,
        y = num_app_gen.values
)]

plotly.offline.iplot(data)
```

# Relation between app rating and app price

```python
trace0 = go.Box(
    # Data for paid apps
    y=appdf[appdf['price'] == 0]['rating_count_tot'],
    name = 'free'
)

trace1 = go.Box(
    # Data for free apps
    y=appdf[appdf['price']>0]['rating_count_tot'] ,
    name = 'paid'
)

layout = go.Layout(
    title = "Total rating of paid apps vs. free apps",
    yaxis = dict(
        type = 'log',
        autorange = True
    )
)

# Add trace0 and trace1 to a list for plotting
data = [trace0, trace1]
plotly.offline.iplot({'data': data, 'layout': layout})
```

# Relation between app genre and app price

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

# Select a few popular app categories
popular_app = appdf[appdf.prime_genre.isin(['Games', 'Entertainment', 'Education',
                                            'Photo & Video', 'Utilities', 'Health & Fitness'])]

# Examine the price trend by plotting Price vs Category
ax = sns.stripplot(x = popular_app['price'], y = popular_app['prime_genre'], jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')

# Apps whose Price is greater than 200
apps_above_200 = popular_app[['prime_genre', 'track_name', 'price']][popular_app['price'] > 200]
apps_above_200
```

```python
app_max50=popular_app[popular_app['price']<51]

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)
ax = sns.stripplot(x = app_max50['price'], y = app_max50['prime_genre'], jitter=True, linewidth=1)
ax.set_title('App pricing max 50 USD')
```
